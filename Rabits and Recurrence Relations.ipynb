```python
def rabbits(n, k):
    if n == 1 or n == 2:
        return 1
    
    f1, f2 = 1, 1
    
    for i in range(3, n + 1):
        f3 = f2 + k * f1
        f1, f2 = f2, f3
    
    return f2


with open("rosalind_fib.txt", "r", encoding="utf-8") as f:
    data = f.read().strip().split()

# convertir a enteros
n = int(data[0])
k = int(data[1])


print(rabbits(n, k))
```
