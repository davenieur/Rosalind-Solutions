```python
with open("rosalind_revc.txt", "r", encoding="utf-8") as f:
    dna = f.read().replace("\n", "")

reversed_dna = dna[::-1]

tabla = str.maketrans({
    "A": "T",
    "C": "G",
    'G': 'C',
    'T': 'A'
})

reversed_complement = reversed_dna.translate(tabla)

print(reversed_complement)
```
